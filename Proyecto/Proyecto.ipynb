### 2.6 Limpiar

Luego de ver nulos y duplicados se decide aplicar la solución más directa y sencilla, eliminando las filas con todos los atributos duplicados y eliminar las filas que contienen nulos. Cabe recalcar que las filas con nulos existen porque no tienen ID común entre el dataframe unido(``df_1_2``) y el dataframe que tiene nuevas features sobre emails(``df_email_phone``).

```python
# Eliminar valores duplicados
df = df.drop_duplicates()
# Eliminar filas con valores nulos
df = df.dropna()
```

Luego de limpiar, creamos una nueva columna llamada ``segmentacion_etaria`` que categoriza la edad de los usuarios en diferentes rangos.

```python
df['segmentacion_etaria'] = pd.cut(
    df['customer_age'], 
    bins=[0,18,27,60,100],     
    labels=['Joven','Adulto-Joven','Adulto','Persona Mayor'],
    right=False)
```

## 4. Preprocesamiento

### 4.1 Declarar `ColumnTransformer`

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler, MinMaxScaler, PowerTransformer
```

### 4.2 Transformar datos

```python
# Columnas para transformar con OneHotEncoder
cols_to_oh = [
    'payment_type',
    'employment_status',
    'housing_status',
    'source',
    'device_os',
    # 'device_distinct_emails_8w',
    'segmentacion_etaria',
]

# Columnas para transformar con OrdinalEncoder
cols_to_o = [
    'segmentacion_etaria'
]

ordinal_cat = [
    ["Joven", "Adulto-Joven", "Adulto", "Persona Mayor"],
]


preprocessing_transformer = ("preprocesamiento", ColumnTransformer(
    transformers = [
        ('OneHotEncoder', OneHotEncoder(sparse_output=False), cols_to_oh),
        ("ordinal_encoder", OrdinalEncoder(categories=ordinal_cat), cols_to_o)
    ],
    remainder='passthrough'
))#.set_output(transform="pandas")

pipeline = Pipeline(
    steps=[
        preprocessing_transformer
    ]
).set_output(transform="pandas")
```

```python
df_preprocesado = pipeline.fit_transform(df)
df_preprocesado
```
