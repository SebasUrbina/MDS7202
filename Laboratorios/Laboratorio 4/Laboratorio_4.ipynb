**Hint**

Puede ser 칰til los m칠todos ```.drop_duplicates()``` y ```.dropna()``` como tambi칠n utilizar queries con m치scaras booleanas.


````python
def clean_data(df_grades):
    # Eliminar alumnos duplicados y valores nulos
    df_grades = df_grades.drop_duplicates(subset=['names']).dropna()
    # Filtrar notas fuera de rango (_)
    df_grades = df_grades.loc[(df_grades['math score']<=100) & ((df_grades['reading score']<=100)) & ((df_grades['writing score']<=100))]
    return df_grades.reset_index(drop=True)
````

**Respuesta:**

````python
df_grades = clean_data(df_grades)
df_grades
````

### 1.3.1 Transformaci칩n de notas a "Nota Chilena" [0.5 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci칩n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci칩n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci칩n entregada, cree una funci칩n que le permita pasar a la escala Chilena las notas que se escapan del rango [1,7] y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
  dataframe_in[columna] = dataframe_in[columna].apply(lambda x: round(x*0.06+1,2))
  return dataframe_in
````

````python
df_grades = nota_chilena(df_grades, ['math score','reading score','writing score'])
df_grades.head()
````

Con todas las notas adjuntadas, la prestigiosa Universidad De la Cacha침a necesita obtener el promedio general de cada uno de los alumnos. Para esto se le solicita que calcule el promedio con cada una de las asignaturas y el promedio se anexe como una nueva columna al final del `DataFrame` original. Cabe se침alar que las notas deben ser redondeadas al segundo decimal.

**To-Do:**
- [ ] Crear funci칩n que nos permite obtener el promedio de cada uno de los alumnos y anexar esta informaci칩n en el `DataFrame` original.


**Respuesta:**

````python
def alumnos_promedio(dataframe_in):
  # calculamos el promedio de los scores
  dataframe_in['GPA'] = dataframe_in[['math score','reading score','writing score']].mean(axis=1).round(2)
  return dataframe_in
````

### 1.3.3 Reprobados y Destacados [1.5 Punto, 0.5 Cada DataFrame]

Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o y los que han tenido desempe침o de destacado. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar치 destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci칩n que permita obtener tres Dataframes: 
- El primero deber치 poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser치 un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a침o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    # reprobados
    dataframe1 = dataframe_in.loc[dataframe_in['GPA']<4.0].reset_index(drop=True)
    
    # reprobados con R en columna GPA
    dataframe2 = dataframe_in.copy()
    dataframe2['GPA'] = dataframe2['GPA'].apply(lambda x: 'R' if x < 4.0 else x)

    # destacados
    dataframe3 = dataframe_in.loc[(dataframe_in['math score']>5.8) & (dataframe_in['reading score']>5.8) & (dataframe_in['writing score']>5.8)].sort_values('GPA', ascending=False).reset_index(drop=True)
    
    return (dataframe1, dataframe2, dataframe3)
````

````python
dataframe1, dataframe2, dataframe3 = reprobados_destacados(df_grades)
dataframe1
````

```
- Para generar el primer dataframe se utiliz칩 el m칠todo de pandas `.loc` para filtrar los los promedios (GPA) menores a 4.0
- Para el segundo dataframe se realiz칩 una copia del dataframe de entrada y se sobreescribri칩 la columna `GPA` utilizando el m칠todo `.apply` usando una funci칩n lambda.
- Para el tercer dataframe se utiliz칩 el m칠todo `.loc` para filtrar los alumnos cuyo promedio es superior a 5.8 en matem치ticas, lectura y escritura y se us칩 el m칠todo `.sort_values` para ordenar los valores en funci칩n de la columna GPA
``` 
